<p style= "text-align: justify;"> La enfermedad por coronavirus (COVID-19) es una enfermedad infecciosa causada por un coronavirus descubierta en el 2019. La mayoría de las personas infectadas con el virus COVID-19 experimentarán una enfermedad respiratoria leve a moderada y se recuperarán sin necesidad de un tratamiento especial. Las personas mayores y aquellas con problemas médicos subyacentes como enfermedades cardiovasculares, diabetes, enfermedades respiratorias crónicas y cáncer tienen más probabilidades de desarrollar una enfermedad grave.

<p style= "text-align: justify;"> Durante todo el transcurso de la pandemia, uno de los principales problemas a los que se han enfrentado los proveedores de atención médica es la escasez de recursos médicos y un plan adecuado para distribuirlos de manera eficiente. En estos tiempos difíciles, poder predecir qué tipo de recurso podría necesitar una persona en el momento de dar positivo en la prueba o incluso antes de eso será de gran ayuda para las autoridades, ya que podrían obtener y organizar los recursos necesarios para salvar la vida de ese paciente.


**<p style= "text-align: left;"> OBJETIVO:** </P>
<p style= "text-align: justify;"> El objetivo principal de este proyecto es construir un modelo de aprendizaje automático que, dado el síntoma, el estado y el historial médico actuales de un paciente con Covid-19, prediga si el paciente tiene alto riesgo o no de muerte. </p>

1. **sexo:** 1 para mujer y 2 para hombre.

2. **edad:** del paciente.

3. **clasificación:** Hallazgos de la prueba de covid. Los valores de 1 a 3 significan que el paciente fue diagnosticado con covid en diferentes
grados. 4 o más significa que el paciente no es portador de covid o que la prueba no es concluyente.

4. **tipo de paciente:** Tipo de atención que recibió el paciente en la unidad. 1 para regreso a casa y 2 para hospitalización.

5. **neumonía:** Si el paciente ya tenía inflamación de los sacos aéreos o no.

6. **embarazo:** Si la paciente está embarazada o no.

7. **diabetes:** Si el paciente tiene diabetes o no.

8. **copd:** Indica si el paciente tiene enfermedad pulmonar obstructiva crónica o no.

9. **asma:** Si el paciente tiene asma o no.

10. **inmsupr:** Si el paciente está inmunodeprimido o no.

11. **hipertensión:** Si el paciente tiene hipertención o no.

12. **cardiovascular:** Si el paciente tiene una enfermedad relacionada con el corazón o los vasos sanguineos.

13. **renal cronica:** Si el paciente tiene renal cronica o no.

14. **otra enfermedad:** Si el paciente tiene otra enfermedad o no.

15. **obesidad:** Si el paciente es obeso o no.

16. **tabaco:** Si el paciente es consumidor de tabaco.

17. **usmr:** Indica si el paciente fue atendido en unidades médicas de primer (1er), segundo (2do) o tercer (3er) nivel.

18. **unidad médica:** tipo de institución del Sistema Nacional de Salud que le brindó la atención.

19. **intubado** si el paciente estaba conectado al respirador.

20. **icu:** Indica si el paciente había estado ingresado en una Unidad de Cuidados Intensivos (ICU en ingles)

21. **fecha de fallecimiento:** si el paciente falleció indique la fecha de fallecimiento, y 9999-99-99 en caso contrario.

### CAMBIO EN LOS VALORES DE LA TABLA:
Se cambiará algunos valores de la tabla para poder utilizarlos, como sería la fecha y elimnar aquellas columnas que no aportan verdaderos valores al planteamiento del problema por escases de datos.

```python
# IMPRIMIMOS TODAS LAS VARIABLES QUE CONTEMPLAMOS
df.columns
```

```python
# DEJAMOS SOLO LAS ENFERMADES EN LA TABLA PARA HACER UN ANALISIS.
cols = ['PNEUMONIA','DIABETES', 'COPD', 'ASTHMA', 'INMSUPR','HIPERTENSION',
        'OTHER_DISEASE', 'CARDIOVASCULAR', 'OBESITY','RENAL_CHRONIC', 'TOBACCO']
for col in cols :
    df = df[(df[col] == 1)|(df[col] == 2)]
```

```python
df.head(10)  #IMPRIMIMOS TABLA CON SOLO 10 VALORES DE HILERA
```

**TRANSFORMAREMOS LA FECHA DE MUERTE:**
- "1" (PARA AFIRMAR QUE MURIÓ) 
- "2" (PARA AFIRMAR QUE VIVE), SON AQUELLOS QUE NO TIENEN FECHA

```python
df['DATE_DIED'] = [2 if row =='9999-99-99' else 1 for row in df['DATE_DIED']]
```

**HEATMAP Y COMANDO CORR()**

Procedemos a imprimir un HEATMAP para encontrar la relación entre cada valor de las tablas.

```python
#CODIGO QUE VENÍA EN LA CLASE DE LA MAESTRA.
def corr_matrix(df): #matriz de correlación
    correlation = df.corr()
    plt.figure(figsize=(20, 15))
    sns.heatmap(correlation, annot=True, cbar=True, cmap="Blues")
    plt.title('TABLA DE CORRELACIÓN ENTRE CARACTERÍSTICAS DE LOS DATOS', color = 'BLACK', fontsize = 20) #TITULO
```

```python
corr_matrix(df)
```

### MEDIA TRUNCADA

```python
def media_truncada(datos, proportiontocut):
    # Ordenar los datos
    datos_ordenados = np.sort(datos)    
    # Calcular el número de elementos a eliminar de cada extremo
    n = len(datos)
    cortar = int(n * proportiontocut)   
    # Eliminar los elementos de los extremos
    datos_truncados = datos_ordenados[cortar:n-cortar]    
    # Calcular la media de los datos truncados
    media_truncada = np.mean(datos_truncados)    
    return media_truncada, datos_truncados
```

```python
# Calcular la media truncada (por ejemplo, eliminando el 10% de los valores más altos y más bajos)
media2, datos_truncados = media_truncada(valores, proportiontocut=0.05)
media2
```

```python
# Graficar el histograma
plt.hist(datos_truncados, bins=20, edgecolor='black')  # Añadir borde a las barras para mejor visualización
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("HISTOGRAMA DE LA MEDIA TRUNCADA DE LOS PACIENTES DE COVID-19")
#--------------------------------------------------------------------------------------------------------------
# Añadir una línea vertical en la media truncada
plt.axvline(media2, color='red', linestyle='dashed', linewidth=1)
# Obtener los límites del eje y
min_ylim, max_ylim = plt.ylim()
# Añadir texto para la media truncada
plt.text(media2 * 1.1, max_ylim * 0.9, f'Media truncada: {media2:.2f}', color='red')
#--------------------------------------------------------------------------------------------------------------
# Mostrar la figura
plt.show()
```

**CONCLUSIONES:**

### MODA.

```python
def calcular_moda(valores):
    # Crear un diccionario para contar la frecuencia de cada valor
    frecuencia = {}
    for valor in valores:
        if valor in frecuencia:
            frecuencia[valor] += 1
        else:
            frecuencia[valor] = 1    
    # Encontrar la frecuencia máxima
    max_frecuencia = max(frecuencia.values())    
    # Encontrar todos los valores con la frecuencia máxima
    modas = [key for key, value in frecuencia.items() if value == max_frecuencia]    
    return modas
```

```python
moda = calcular_moda(valores)
print(f"La moda es: {moda}")
```

```python
# Graficar el histograma de los datos
plt.hist(valores, bins=20, edgecolor='black')  # Añadir borde a las barras para mejor visualización
plt.xlabel("EDAD")
plt.ylabel("FRECUENCIA")
plt.title("HISTOGRAMA DE LA EDAD CON LA MODA")
#-------------------------------------------------------------------------------------------------------
# Añadir líneas verticales para cada moda
for m in moda:
    plt.axvline(m, color='red', linestyle='dashed', linewidth=1)
# Obtener los límites del eje y
min_ylim, max_ylim = plt.ylim()
# Añadir texto para la moda
for m in moda:
    plt.text(m, max_ylim * 0.9, f'Moda: {m}', rotation=0, verticalalignment='center', color='red')
# Mostrar la figura
plt.show()
```

### MEDIANA.

```python
def calcular_mediana(valores):
    # Ordenar los valores
    valores_ordenados = sorted(valores)   
    # Calcular el número de elementos
    n = len(valores_ordenados)    
    # Encontrar la mediana
    if n % 2 == 1:
        # Si el número de elementos es impar, la mediana es el valor del medio
        mediana = valores_ordenados[n // 2]
    else:
        # Si el número de elementos es par, la mediana es el promedio de los dos valores del medio
        mediana = (valores_ordenados[n // 2 - 1] + valores_ordenados[n // 2]) / 2    
    return mediana
```

```python
mediana = calcular_mediana(valores)
print(f"La mediana es: {mediana}")
```

```python
# Graficar el histograma de los datos
plt.hist(valores, bins=20, edgecolor='black')  # Añadir borde a las barras para mejor visualización
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Histograma de los valores con la mediana")

# Añadir una línea vertical en la mediana
plt.axvline(mediana, color='red', linestyle='dashed', linewidth=1)

# Obtener los límites del eje y
min_ylim, max_ylim = plt.ylim()

# Añadir texto para la mediana
plt.text(mediana * 1.01, max_ylim * 0.9, f'Mediana: {mediana}', rotation=0, verticalalignment='center', color='red')

# Mostrar la figura
plt.show()
```

```python
# VARIANZA.
mu = media_aritmetica
def varianza(datos):
    columns = len(datos)
    sumatory = sum((datos-mu)*(datos-mu))
    return round(sumatory/columns,2)
varianza(valores)
```

### DESVIACIÓN ESTANDAR.

```python
def calcular_desviacion_estandar(valores):
    # Calcular la media aritmética
    media_aritmetica = sum(valores) / len(valores)    
    # Calcular las desviaciones cuadráticas de la media
    desviaciones_cuadraticas = [(x - media_aritmetica) ** 2 for x in valores]    
    # Calcular la varianza
    varianza = sum(desviaciones_cuadraticas) / len(valores)   
    # Calcular la desviación estándar
    desviacion_estandar = math.sqrt(varianza)   
    return desviacion_estandar
```

```python
# Calcular la desviación estándar
desviacion_estandar = calcular_desviacion_estandar(valores)
print(f"La desviación estándar es: {desviacion_estandar}")
```

```python
# Graficar el histograma de los datos
plt.hist(valores, bins=20, edgecolor='black')  # Añadir borde a las barras para mejor visualización
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Histograma de los valores con la desviación estándar")
#--------------------------------------------------------------------------------------------------------------
# Añadir una línea vertical en la media
plt.axvline(media_aritmetica, color='yellow', linestyle='dashed', linewidth=1)
# Añadir líneas verticales para la desviación estándar
plt.axvline(media_aritmetica + desviacion_estandar, color='red', linestyle='dashed', linewidth=1)
plt.axvline(media_aritmetica - desviacion_estandar, color='red', linestyle='dashed', linewidth=1)
# Obtener los límites del eje y
min_ylim, max_ylim = plt.ylim()
#---------------------------------------------------------------------------------------------------------------
# Añadir texto para la desviación estándar
plt.text(media_aritmetica + desviacion_estandar, max_ylim * 0.9, f'+1σ: {media_aritmetica + desviacion_estandar:.2f}', rotation=0, verticalalignment='center', color='r')
plt.text(media_aritmetica - desviacion_estandar, max_ylim * 0.9, f'-1σ: {media_aritmetica - desviacion_estandar:.2f}', rotation=0, verticalalignment='center', color='r')
# Añadir leyenda
plt.legend()
# Mostrar la figura
plt.show()
```

- Las desviaciones estándar **más pequeñas** indican que la mayoría de los valores de los datos están próximos a la media muestral.
- Las desviaciones estándar **más grandes** indican que los valores de los datos están más dispersos y que algunos valores están más alejados de la media muestral.

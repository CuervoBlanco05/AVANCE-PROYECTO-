# **<p style= "text-align: center;"> <font color='blue'>CONJUNTO DE DATOS SOBRE COVID-19. </font> </p>**

<p style= "text-align: justify;"> La enfermedad por coronavirus (COVID-19) es una enfermedad infecciosa causada por un coronavirus descubierta en el 2019. La mayoría de las personas infectadas con el virus COVID-19 experimentarán una enfermedad respiratoria leve a moderada y se recuperarán sin necesidad de un tratamiento especial. Las personas mayores y aquellas con problemas médicos subyacentes como enfermedades cardiovasculares, diabetes, enfermedades respiratorias crónicas y cáncer tienen más probabilidades de desarrollar una enfermedad grave.

<p style= "text-align: justify;"> Durante todo el transcurso de la pandemia, uno de los principales problemas a los que se han enfrentado los proveedores de atención médica es la escasez de recursos médicos y un plan adecuado para distribuirlos de manera eficiente. En estos tiempos difíciles, poder predecir qué tipo de recurso podría necesitar una persona en el momento de dar positivo en la prueba o incluso antes de eso será de gran ayuda para las autoridades, ya que podrían obtener y organizar los recursos necesarios para salvar la vida de ese paciente.


**<p style= "text-align: left;"> OBJETIVO:** </P>
<p style= "text-align: justify;"> El objetivo principal de este proyecto es construir un modelo de aprendizaje automático que, dado el síntoma, el estado y el historial médico actuales de un paciente con Covid-19, prediga si el paciente tiene alto riesgo o no. </p>

1. **sexo:** 1 para mujer y 2 para hombre.

2. **edad:** del paciente.

3. **clasificación:** Hallazgos de la prueba de covid. Los valores de 1 a 3 significan que el paciente fue diagnosticado con covid en diferentes
grados. 4 o más significa que el paciente no es portador de covid o que la prueba no es concluyente.

4. **tipo de paciente:** Tipo de atención que recibió el paciente en la unidad. 1 para regreso a casa y 2 para hospitalización.

5. **neumonía:** Si el paciente ya tenía inflamación de los sacos aéreos o no.

6. **embarazo:** Si la paciente está embarazada o no.

7. **diabetes:** Si el paciente tiene diabetes o no.

8. **copd:** Indica si el paciente tiene enfermedad pulmonar obstructiva crónica o no.

9. **asma:** Si el paciente tiene asma o no.

10. **inmsupr:** Si el paciente está inmunodeprimido o no.

11. **hipertensión:** Si el paciente tiene hipertención o no.

12. **cardiovascular:** Si el paciente tiene una enfermedad relacionada con el corazón o los vasos sanguineos.

13. **renal cronica:** Si el paciente tiene renal cronica o no.

14. **otra enfermedad:** Si el paciente tiene otra enfermedad o no.

15. **obesidad:** Si el paciente es obeso o no.

16. **tabaco:** Si el paciente es consumidor de tabaco.

17. **usmr:** Indica si el paciente fue atendido en unidades médicas de primer (1er), segundo (2do) o tercer (3er) nivel.

18. **unidad médica:** tipo de institución del Sistema Nacional de Salud que le brindó la atención.

19. **intubado** si el paciente estaba conectado al respirador.

20. **icu:** Indica si el paciente había estado ingresado en una Unidad de Cuidados Intensivos (ICU en ingles)

21. **fecha de fallecimiento:** si el paciente falleció indique la fecha de fallecimiento, y 9999-99-99 en caso contrario.

### CAMBIO EN LOS VALORES DE LA TABLA:
Se cambiará algunos valores de la tabla para poder utilizarlos, como sería la fecha y elimnar aquellas columnas que no aportan verdaderos valores al planteamiento del problema por escases de datos.

```python
# DEJAMOS SOLO LAS ENFERMADES EN LA TABLA PARA HACER UN ANALISIS.
cols = ['PNEUMONIA','DIABETES', 'COPD', 'ASTHMA', 'INMSUPR','HIPERTENSION',
        'OTHER_DISEASE', 'CARDIOVASCULAR', 'OBESITY','RENAL_CHRONIC', 'TOBACCO']
for col in cols :
    df = df[(df[col] == 1)|(df[col] == 2)]
```

## 2. GRAFICAS DE DATOS.

### TABLA DE CORRELACIÓN ENTRE CARACTERÍSTICAS DE LOS DATOS.
**HEATMAP**

```python
#CODIGO QUE ME ENCONTRÉ.
plt.figure(figsize=(20, 15))
plt.title('TABLA DE CORRELACIÓN ENTRE CARACTERÍSTICAS DE LOS DATOS', fontsize=14, color='black')  #TITULO DEL GRAFICO 
sns.heatmap(df.drop(columns=['DATE_DIED']).corr(), annot=True, fmt='.2f', cmap='Blues', square=True, linewidths=.5) #cmap es la parte para cambiar color
plt.show()   
```

```python
#CODIGO QUE ME ENCONTRÉ EN INTERNET.
fig, ax = plt.subplots(figsize=(20, 15))
mask=np.triu(np.ones_like(df.drop(columns=['DATE_DIED']).corr()))
sns.heatmap(df.drop(columns=['DATE_DIED']).corr(), mask = mask, annot = True, cmap = "Blues", vmin = -1, vmax = 1)
plt.title('TABLA DE CORRELACIÓN ENTRE CARACTERÍSTICAS DE LOS DATOS', color = 'BLUE', fontsize = 20)
plt.show()
```

```python
#CODIGO QUE VENÍA EN LA CLASE DE LA MAESTRA.
def corr_matrix(df): #matriz de correlación
    correlation = df.drop(columns=['DATE_DIED']).corr()
    plt.figure(figsize=(20, 15))
    sns.heatmap(correlation, annot=True, cbar=True, cmap="Blues")
    plt.title('TABLA DE CORRELACIÓN ENTRE CARACTERÍSTICAS DE LOS DATOS', color = 'BLUE', fontsize = 20) #TITULO
```

```python
corr_matrix(df)
```

se quita del _**HEATMAP**_ _DATE_DIED_ pues lo que nos insteresa es la relación entre los valores de enfermedad.

```python
vivo_muerto= df.groupby(df['PNEUMONIA'])["DATE_DIED"].value_counts(normalize=False).unstack()
plt.figure(figsize=(10, 5))
vivo_muerto.plot(kind='bar',stacked=False)
plt.xlabel(df['PNEUMONIA'])
plt.ylabel('Count')
plt.title('RELACIÓN VARIABLE-MUERTOS')
plt.legend(title='Datos de muertos')
plt.show()
```

```python
vivo_muerto= df.groupby(df['DIABETES'])["DATE_DIED"].value_counts(normalize=False).unstack()
plt.figure(figsize=(10, 5))
vivo_muerto.plot(kind='bar',stacked=False)
plt.xlabel(df['DIABETES'])
plt.ylabel('Count')
plt.title('RELACIÓN VARIABLE-MUERTOS')
plt.legend(title='Datos de muertos')
plt.show()
```

```python
vivo_muerto= df.groupby(df['COPD'])["DATE_DIED"].value_counts(normalize=False).unstack()
plt.figure(figsize=(10, 5))
vivo_muerto.plot(kind='bar',stacked=False)
plt.xlabel(df['COPD'])
plt.ylabel('Count')
plt.title('RELACIÓN VARIABLE-MUERTOS')
plt.legend(title='Datos de muertos')
plt.show()
```
